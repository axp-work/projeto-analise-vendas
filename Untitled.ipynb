```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Bibliotecas importadas com sucesso!")
```

```python
#criar dados ficticios de vendas
dados = {
    'Produtos': ['Camiseta', 'Calça', 'Tênis', 'Boné', 'Meia'],
    'Quantidade_Vendida': [150, 80,200, 120, 300],
    'Preco_Unitario': [29.90, 89.90, 159.90, 19.90, 9.90],
    'Categoria': ['Vestuário', 'Vestuário', 'Calçados', 'Acessórios', 'Vestuário']
}

#Converter para dataframe (Tabela).
df = pd.DataFrame (dados)

#Exibir a tabela
df
```

```python
#Criar coluna de faturamento (Quantidade x Preço)
df['Faturamento'] = df['Quantidade_Vendida'] * df['Preco_Unitario']

#Exibir a tabela atualizada
df
```

```python
#Qual produto gerou mais faturamento?
df[df['Faturamento'] == df['Faturamento'].max()]
```

```python
#Faturamento total por categoria
faturamento_categoria = df.groupby('Categoria')['Faturamento'].sum()

#exibir o resultado
faturamento_categoria
```

```python
#Configurar gráfico
plt.figure(figsize=(8,5))

#criar gráfico de barras
sns.barplot(x = faturamento_categoria.index, y = faturamento_categoria.values)

#títulos e rótulos
plt.title('Faturamento Total por categoria', fontsize = 14)

#exibição do gráfico
plt.show()
```

```python
#Salvando arquivo
plt.figure (figsize=(8, 5))
sns.barplot (x = faturamento_categoria.index, y = faturamento_categoria.values)
plt.title ('Faturamento total por categoria', fontsize=14)
plt.xlabel ('Categoria')
plt.ylabel ('Faturamento (R$)')

#salvar imagem
plt.savefig('Faturamento por categoria.png', dpi=100, bbox_inches='tight')
plt.show()
```
